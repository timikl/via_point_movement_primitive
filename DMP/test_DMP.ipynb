```python
import numpy as np
import matplotlib.pyplot as plt 
%matplotlib inline

from dmp import DMP
```

```python
input_trajectory = np.sin(np.linspace(0,10,1000))
input_trajectory = input_trajectory.reshape(input_trajectory.shape[0], 1)
print(input_trajectory.shape)

dmp = DMP(pos_data = input_trajectory, time = 1/100. )
```

```python
output_trajectory, t = dmp.decode()
```

```python
plt.plot(input_trajectory)
plt.plot(output_trajectory)
```
